**Dt_Customer(고객 가입일) : 가입일 확인**
* 데이터 타입 시계열로 변경



```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
dates = []

for i in df['Dt_Customer']:
  dates.append(i)


print(f'가장 최근에 가입한 고객이 가입한 날짜 : {max(dates)}')
print(f'고객들 중 가장 먼저 가입한 고객의 날짜 : {min(dates)}')
```

```python
#  Period of use
#  최근 가입일을 기준으로 고객들의 '가입 후 이용기간'을 확인한다.
days = []
latest_sign_up = max(dates)

for i in dates:
  delta = latest_sign_up - i
  days.append(delta)
days[0]


# timedelta 타입의 데이터를 정수변환하기 위해서 .days 메소드를 이용하여 정수만 추출한다.
days_int = []
for i in range(0, len(days)): 
  days_int.append(days[i].days)

days_int[0]
```

```python
# 이용기간을 나타내는 컬럼을 하나 생성
df['Using_Period'] = days_int
```

####  데이터 엔지니어링
>  주어진 변수들을 이용하여 분석에 도움이 될 것 같은 변수들을 추가로 생성하기.
 

```python
def Data_Engineering(df):


  #1. ------------ 새롭게 생성된 변수는 다음과 같습니다.

  # Age : 나이를 나타내는 변수 생성
  df['Age'] = 2022 - df['Year_Birth']
  
  # Children : 두 타입의 자녀의 수를 더하여 자녀의 수를 하나로 나타낼 것.
  df['Children'] = df['Kidhome'] + df['Teenhome']

  # Spent_All : 손님들의 구매이력을 모두 더하여 총 판매금액을 나타낼 것.
  df['Spent_All'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds'] 

  # PurchaseNumAll : 구매횟수와 관련된 컬럼을 모두 더하여 총 구매 횟수를 나타내기.
  df['PurchaseNumAll'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'] + df['NumWebVisitsMonth']
    
  # PurDeal_PurAll_ratio : 할인 구매의 비율 (할인에 의한 구매횟수 / 총 구매 횟수) 
  df['PurDeal_PurAll_ratio'] = df['NumDealsPurchases'] / df['PurchaseNumAll']
  
  # Count_Promos : 프로모션에서 동의한 수를 나타낼 것. (6번의 프로모션을 동의했다면 값은 6일 것.)
  df['Count_Promos'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response']
    
  # Marital_status 카테고리 변경 : 변수이름을 Living_with 으로 변경 및 5개의 범주를 맵핑
  df.rename(columns = {'Marital_Status' : 'Living_with'}, inplace = True)
  df.loc[(df['Living_with'] == 'Married') | (df['Living_with'] == 'Together'), 'Living_with'] = 'Partner'
  df.loc[df['Living_with'] != 'Partner', 'Living_with'] = 'Alone'
  
  # Family_Size : 위에서 생성한 Living_with과 children을 더하여 총 가족(구성원) 수를 구합니다.
  df['Family_Size'] = df['Children'] + df['Living_with'].replace({'Partner':2, 'Alone':1})
  
  # 자녀의 수가 0보다 많으면 부모, 없다면
  df['Parent'] = np.where(df['Children'] > 0, 1, 0)



  # Education 카테고리 변경 : 특정 카테고리 데이터를 묶어준다.
  df['Education'] = df['Education'].replace({'2n Cycle' : 'Post Graduate',
                                             'Master' : 'Post Graduate',
                                             'PhD' : 'Post Graduate'})
  
```

```python
# 생성한 함수 적용
Data_Engineering(df)
```
